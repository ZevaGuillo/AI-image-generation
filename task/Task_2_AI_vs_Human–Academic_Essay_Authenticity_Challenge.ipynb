# Balancear por SMOTE

```python
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer

# Asignar X e Y
X_train = df_train['essay']
Y_train = df_train['label']

# 1. Create a TF-IDF vectorizer
vectorizer = TfidfVectorizer()

# 2. Fit the vectorizer to your training data and transform it
X_train_tfidf = vectorizer.fit_transform(X_train)

# 3. Now apply SMOTE to the numerical representation
smote = SMOTE(random_state=42)
X_smote, y_smote = smote.fit_resample(X_train_tfidf, Y_train)

# Print the shape of the original and resampled data
print("Original dataset shape:", X_train_tfidf.shape, Y_train.shape)
print("Resampled dataset shape:", X_smote.shape, y_smote.shape)

# Print the value counts of the target variable before and after SMOTE
print("\nOriginal target variable distribution:")
print(Y_train.value_counts())
print("\nResampled target variable distribution:")
print(y_smote.value_counts())

Y_train
```
